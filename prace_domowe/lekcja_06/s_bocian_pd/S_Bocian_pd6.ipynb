```python
# Zadanie 3

metryki = {
    "accuracy": 0.94,
    "precision": 0.91,
    "recall": 0.89,
    "f1_score": 0.90
}


print("Najwyższa wartość:", max(metryki.values()))
print("Najniższa wartość:", min(metryki.values()))

srednia = sum(metryki.values()) / len(metryki)
print(f"Średnia metryka: {srednia:.3f}")

print("Posortowane klucze:", sorted(metryki.keys()))

print("Posortowane wartości:", sorted(metryki.values()))
```

```python
# Zadanie 5

grupa_a = {'python', 'pandas', 'numpy', 'sql', 'tableau'}
grupa_b = {'python', 'r', 'sql', 'ggplot2', 'dplyr'}

# Suma |
print("Suma:", grupa_a | grupa_b)
print("Suma:", grupa_a.union(grupa_b))

# Iloczyn &
print("Iloczyn:", grupa_a & grupa_b)
print("Iloczyn:", grupa_a.intersection(grupa_b))

# Różnica -
print("Różnica:", grupa_a - grupa_b)
print("Różnica:", grupa_a.difference(grupa_b))

# Różnica symetryczna ^
print("Różnica symetryczna:", grupa_a ^ grupa_b)
print("Różnica symetryczna:", grupa_a.symmetric_difference(grupa_b))

# Czy zbiory są rozłączne?
print("Czy zbiory są rozłączne?", grupa_a.isdisjoint(grupa_b))
```

```python
# Zadanie 6

rekord = ('setosa', 5.1, 3.5, 1.4, 0.2)

# Rozpakowanie
gatunek, sl, sw, pl, pw = rekord

# Utworzenie słownika
rekord_slownik = {
    'species': gatunek,
    'sepal_length': sl,
    'sepal_width': sw,
    'petal_length': pl,
    'petal_width': pw
}

# Wypisanie w formacie tabelarycznym
print(f"{'species':13}: {gatunek}")

for klucz in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    print(f"{klucz:13}: {rekord_slownik[klucz]:5.1f}")

# Suma cech liczbowych
suma_cech = sum(rekord[1:])

# \n = nowa linia
print(f"\nSuma cech liczbowych: {suma_cech:.1f}")
```
