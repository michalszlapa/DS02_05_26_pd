```python
# Zadanie 3. Metryki modelu
metryki = {
    "accuracy": 0.94,
    "precision": 0.91,
    "recall": 0.89,
    "f1_score": 0.90
}

print("Najwyższa wartość:")
print(max(metryki.values()))

print("Najniższa wartość:")
print(min(metryki.values()))

srednia = sum(metryki.values()) / len(metryki)

print("Średnia:")
print(round(srednia, 3))

print("Klucze alfabetycznie:")
print(sorted(metryki))

print("Wartości rosnąco:")
print(sorted(metryki.values()))
```

```python
# Zadanie 5. Operacje na dwóch zbiorach
grupa_a = {'python', 'pandas', 'numpy', 'sql', 'tableau'}
grupa_b = {'python', 'r', 'sql', 'ggplot2', 'dplyr'}

print("1. Wszystkie narzędzia:")
print(grupa_a | grupa_b)
print(grupa_a.union(grupa_b))

print("2. Wspólne narzędzia:")
print(grupa_a & grupa_b)
print(grupa_a.intersection(grupa_b))

print("3. Tylko w grupie A:")
print(grupa_a - grupa_b)
print(grupa_a.difference(grupa_b))

print("4. Tylko w jednej grupie:")
print(grupa_a ^ grupa_b)
print(grupa_a.symmetric_difference(grupa_b))

print("5. Czy zbiory są rozłączne?")
print(grupa_a.isdisjoint(grupa_b))
```
