Zadanie 5. Operacje na dwóch zbiorach
Dane są dwa zbiory narzędzi używanych przez dwie grupy analityków:
grupa_a = {'python', 'pandas', 'numpy', 'sql', 'tableau'}
grupa_b = {'python', 'r', 'sql', 'ggplot2', 'dplyr'}
Używając zarówno operatorów ( | , & , - , ^ ), jak i odpowiadających im metod, wypisz:
Sumę zbiorów (wszystkie narzędzia).
Wspólne narzędzia (iloczyn).
Narzędzia tylko w grupie A (różnica grupa_a - grupa_b ).
Narzędzia tylko w jednej z grup (różnica symetryczna).
Czy zbiory są rozłączne? ( .isdisjoint() )

```python
grupa_a = {'python', 'pandas', 'numpy', 'sql', 'tableau'}
grupa_b = {'python', 'r', 'sql', 'ggplot2', 'dplyr'}
wszystkie_narzedzia = grupa_a | grupa_b
print("Wszystkie narzędzia: ", wszystkie_narzedzia)
wspolne_narzedzia = grupa_a & grupa_b
print("Wspólne narzędzia: ", wspolne_narzedzia)
narzedzia_tylko_A = grupa_a - grupa_b
print("Narzedzia tylko w grupie A: ", narzedzia_tylko_A)
narzedzia_jedna_grupa = grupa_a ^ grupa_b
print("Narzędzia tylko w jednej z grup: ", narzedzia_jedna_grupa)
print(grupa_a.isdisjoint(grupa_b)) #Zbiory nie są rozłączne, posiadają wspólne elementy
```

Zadanie 12. Pełne studium pojedynczego rekordu
Dany jest rekord datasetu Iris w postaci krotki:
rekord = ('setosa', 5.1, 3.5, 1.4, 0.2)
Pierwszy element to nazwa gatunku, kolejne cztery – kolejno sepal_length , sepal_width ,
petal_length , petal_width .
Rozpakuj rekord do zmiennych: gatunek, sl, sw, pl, pw .
Utwórz słownik rekord_slownik , w którym kluczami są 'species' , 'sepal_length' ,
'sepal_width' , 'petal_length' , 'petal_width' , a wartościami odpowiednie pola krotki.
Wypisz w formacie tabelarycznym:
species : setosa
sepal_length : 5.1
sepal_width : 3.5
petal_length : 1.4
petal_width : 0.2
(nazwa klucza wyrównana do lewej na 13 znakach, liczba do prawej na 5 z jednym miejscem po przecinku).
Używając sum i slajsu krotki, wypisz sumę wszystkich czterech cech liczbowych ( sum(rekord[1:]) ).

```python
rekord = ('setosa', 5.1, 3.5, 1.4, 0.2)
gatunek = rekord[0]
sl = rekord[1]
sw = rekord[2]
pl = rekord[3]
pw = rekord[4]

rekord_slownik = {
    'species': gatunek,
    'sepal_length': sl,
    'sepal_width': sw,
    'petal_length': pl,
    'petal_width': pw
}
print(f"{'species:':<13}  {gatunek:>5}")
print(f"{'sepal_length:':<13}  {sl:>5}")
print(f"{'sepal_width:':<13}  {sw:>5}")
print(f"{'petal_length:':<13}  {pl:>5}")
print(f"{'petal_width:':<13}  {pw:>5}")
print(sum(rekord[1:]))
```
