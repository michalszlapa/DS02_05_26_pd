```python
metryki = {
    "accuracy": 0.94,
    "precision": 0.91,
    "recall": 0.89,
    "f1_score": 0.90
}

print("Najwyższa wartość:", max(metryki.values()))
print("Najniższa wartość:", min(metryki.values()))

srednia = sum(metryki.values()) / len(metryki)
print(f"Średnia: {srednia:.3f}")

print("Klucze alfabetycznie:")
print(sorted(metryki.keys()))

print("Wartości rosnąco:")
print(sorted(metryki.values()))
```

zad 5.

```python
grupa_a = {'python', 'pandas', 'numpy', 'sql', 'tableau'}
grupa_b = {'python', 'r', 'sql', 'ggplot2', 'dplyr'}

# 1. Suma zbiorów
print("Suma zbiorów:")
print(grupa_a | grupa_b)

# 2. Wspólne elementy
print("\nWspólne narzędzia:")
print(grupa_a & grupa_b)

# 3. Tylko w grupie A
print("\nTylko w grupie A:")
print(grupa_a - grupa_b)

# 4. Tylko w jednej grupie
print("\nRóżnica symetryczna:")
print(grupa_a ^ grupa_b)

# 5. Czy zbiory są rozłączne?
print("\nCzy zbiory są rozłączne?")
print(grupa_a.isdisjoint(grupa_b))
```

metody 

```python
print(grupa_a.union(grupa_b))
print(grupa_a.intersection(grupa_b))
print(grupa_a.difference(grupa_b))
print(grupa_a.symmetric_difference(grupa_b))
```

```python
rekord = ('setosa', 5.1, 3.5, 1.4, 0.2)

# 1. Rozpakowanie 
gatunek, sl, sw, pl, pw = rekord

# 2. Utworzenie słownika
rekord_slownik = {
    'species': gatunek,
    'sepal_length': sl,
    'sepal_width': sw,
    'petal_length': pl,
    'petal_width': pw
}

# 3. Wypisanie tabelaryczne
print(f"{'species':<13} : {gatunek}")

for klucz in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    print(f"{klucz:<13} : {rekord_slownik[klucz]:>5.1f}")

# 4. Suma cech liczbowych
suma_cech = sum(rekord[1:])
print("\nSuma cech:", suma_cech)
```
