Zadanie 5. Operacje na dwóch zbiorach

```python
grupa_a = {'python', 'pandas', 'numpy', 'sql', 'tableau'}
grupa_b = {'python', 'r', 'sql', 'ggplot2', 'dplyr'}

print(f"""
      Suma: {grupa_a | grupa_b}
      Iloczyn: {grupa_a & grupa_b}
      Narzędzia A: {grupa_a - grupa_b}
      Narzędzia, które się nie powtarzają: {grupa_a ^ grupa_b}
      Czy zbiory są rozłączne: {grupa_a.isdisjoint(grupa_b)}
""")
```
