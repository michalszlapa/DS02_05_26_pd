# Utwórz słownik metryki
z kluczami: 
* accuracy,
*  precision,
*  recall. 
* f1_score, 


 i wartościami odpowiednio; 
* 0.94
* 0.91
* 0.89
* 0.90

Następnie używając funkcji wbudowanych:
* min
* max
* sum
* len
* sorted):1.

Wypisz najwyższą i najniższą wartość metryki.2.
Wypisz średnią metrykę z dokładnością do 3 miejsc po przecinku.3.
Wypisz klucze posortowane alfabetycznie (sorted).4.
Wypisz wartości posortowane rosnąco.

```python

metryki = {
    "accuracy": 0.94,
    "precision": 0.91,
    "recall": 0.89,
    "f1_score": 0.90
}


print("Najwyższa wartość:", max(metryki.values()))
print("Najniższa wartość:", min(metryki.values()))

srednia = sum(metryki.values()) / len(metryki)
print(f"Średnia metryka: {srednia:.3f}")

print("Klucze posortowane alfabetycznie:")
print(sorted(metryki.keys()))


print("Wartości posortowane rosnąco:")
print(sorted(metryki.values()))
```

# Operacje na dwóch zbiorach

Dane są dwa zbiory narzędzi używanych przez dwie grupy analityków:

grupa_a

}
Używając

operatorów (

,

,

), jak i odpowiadających im metod, wypisz:
* Sumę zbiorów (wszystkie narzędzia).
* Wspólne narzędzia (iloczyn).3.
* Narzędzia tylko w grupie A (różnica grupa_a - grupa_b).
* Narzędzia tylko w jednej z grup (różnica symetryczna).
* Czy zbiory są rozłączne? (.isdisjoint())

```python

grupa_a = {'python', 'pandas', 'numpy', 'sql', 'tableau'}
grupa_b = {'python', 'r', 'sql', 'ggplot2', 'dplyr'}


print("Suma zbiorów (operator |):")
print(grupa_a | grupa_b)

print("\nSuma zbiorów (metoda union()):")
print(grupa_a.union(grupa_b))


print("\nWspólne narzędzia (operator &):")
print(grupa_a & grupa_b)

print("\nWspólne narzędzia (metoda intersection()):")
print(grupa_a.intersection(grupa_b))

print("\nTylko w grupie A (operator -):")
print(grupa_a - grupa_b)

print("\nTylko w grupie A (metoda difference()):")
print(grupa_a.difference(grupa_b))

print("\nRóżnica symetryczna (operator ^):")
print(grupa_a ^ grupa_b)

print("\nRóżnica symetryczna (metoda symmetric_difference()):")
print(grupa_a.symmetric_difference(grupa_b))


print("\nCzy zbiory są rozłączne?")
print(grupa_a.isdisjoint(grupa_b))
```
