# Zadanie 3. Metryki modelu
Utwórz słownik metryki z kluczami accuracy , precision , recall , f1_score i wartościami
odpowiednio 0.94 , 0.91 , 0.89 , 0.90 . Następnie używając funkcji wbudowanych ( min , max , sum ,
len , sorted ):
1. Wypisz najwyższą i najniższą wartość metryki.
2. Wypisz średnią metrykę z dokładnością do 3 miejsc po przecinku.
3. Wypisz klucze posortowane alfabetycznie ( sorted ).
4. Wypisz wartości posortowane rosnąco.


```python
metryki = {'accuracy': 0.94, 'precision': 0.91, 'recall': 0.89, 'f1_score': 0.90}

print(f"Najwyższa wartość: {max(metryki.values())}")
print(f"Najniższa wartość: {min(metryki.values())}")

srednia = sum(metryki.values()) / len(metryki)
print(f"Średnia metryka: {srednia:.3f}")

print(f"Klucze posortowane: {sorted(metryki.keys())}")
print(f"Wartości posortowane: {sorted(metryki.values())}")
```

# Zadanie 5. Operacje na dwóch zbiorach
Dane są dwa zbiory narzędzi używanych przez dwie grupy analityków:
`grupa_a = {'python', 'pandas', 'numpy', 'sql', 'tableau'}`
`grupa_b = {'python', 'r', 'sql', 'ggplot2', 'dplyr'}`
Używając zarówno operatorów ( | , & , - , ^ ), jak i odpowiadających im metod, wypisz:
1. Sumę zbiorów (wszystkie narzędzia).
2. Wspólne narzędzia (iloczyn).
3. Narzędzia tylko w grupie A (różnica grupa_a - grupa_b ).
4. Narzędzia tylko w jednej z grup (różnica symetryczna).
5. Czy zbiory są rozłączne? ( .isdisjoint() )

```python
grupa_a = {'python', 'pandas', 'numpy', 'sql', 'tableau'}
grupa_b = {'python', 'r', 'sql', 'ggplot2', 'dplyr'}

print(f"Suma (operator |): {grupa_a | grupa_b}")
print(f"Suma (metoda union): {grupa_a.union(grupa_b)}")

print(f"Iloczyn (operator &): {grupa_a & grupa_b}")
print(f"Iloczyn (metoda intersection): {grupa_a.intersection(grupa_b)}")

print(f"Różnica (operator -): {grupa_a - grupa_b}")
print(f"Różnica (metoda difference): {grupa_a.difference(grupa_b)}")

print(f"Różnica symetryczna (operator ^): {grupa_a ^ grupa_b}")
print(f"Różnica symetryczna (metoda symmetric_difference): {grupa_a.symmetric_difference(grupa_b)}")

print(f"Czy zbiory są rozłączne: {grupa_a.isdisjoint(grupa_b)}")
```

# Zadanie 12. Pełne studium pojedynczego rekordu
Dany jest rekord datasetu Iris w postaci krotki:
`rekord = ('setosa', 5.1, 3.5, 1.4, 0.2)`
Pierwszy element to nazwa gatunku, kolejne cztery – kolejno `sepal_length`, `sepal_width`, `petal_length`, `petal_width`.
1. Rozpakuj rekord do zmiennych: gatunek, sl, sw, pl, pw .
2. Utwórz słownik `rekord_slownik`, w którym kluczami są 'species', 'sepal_length', 'sepal_width', 'petal_length', 'petal_width', a wartościami odpowiednie pola krotki.
3. Wypisz w formacie tabelarycznym:
`species : setosa
sepal_length : 5.1
sepal_width : 3.5
petal_length : 1.4
petal_width : 0.2`
(nazwa klucza wyrównana do lewej na 13 znakach, liczba do prawej na 5 z jednym miejscem po przecinku).
1. Używając sum i slajsu krotki, wypisz sumę wszystkich czterech cech liczbowych ( sum(rekord[1:]) )

```python
rekord = ('setosa', 5.1, 3.5, 1.4, 0.2)
gatunek, sl, sw, pl, pw = rekord

rekord_slownik = {
    'species': gatunek,
    'sepal_length': sl,
    'sepal_width': sw,
    'petal_length': pl,
    'petal_width': pw
}

print(f"""{'species':13}: {rekord_slownik['species']:>5}
{'sepal_length':13}: {rekord_slownik['sepal_length']:5.1f}
{'sepal_width':13}: {rekord_slownik['sepal_width']:5.1f}
{'petal_length':13}: {rekord_slownik['petal_length']:5.1f}
{'petal_width':13}: {rekord_slownik['petal_width']:5.1f}""")

suma = sum(rekord[1:])
print(f"Suma cech liczbowych: {suma}")
```
