### Zadanie 5. Operacje na dwóch zbiorach

Dane są dwa zbiory narzędzi używanych przez dwie grupy analityków:  
grupa_a = {'python', 'pandas', 'numpy', 'sql', 'tableau'}  
grupa_b = {'python', 'r', 'sql', 'ggplot2', 'dplyr'}  

Używając zarówno operatorów ( | , & , - , ^ ), jak i odpowiadających im metod, wypisz:  
1. Sumę zbiorów (wszystkie narzędzia).  
2. Wspólne narzędzia (iloczyn).  
3. Narzędzia tylko w grupie A (różnica grupa_a - grupa_b ).  
4. Narzędzia tylko w jednej z grup (różnica symetryczna).  
5. Czy zbiory są rozłączne? ( .isdisjoint() )

```python
grupa_a = {'python', 'pandas', 'numpy', 'sql', 'tableau'}
grupa_b = {'python', 'r', 'sql', 'ggplot2', 'dplyr'}

print(f"Suma zbiorów: {grupa_a | grupa_b}")
print(f"Suma zbiorów: {grupa_a.union(grupa_b)}")
print(f"Wspólna narzędzia to: {grupa_a & grupa_b}")
print(f"Wspólne narzędzia to: {grupa_a.intersection(grupa_b)}")
print(f"Narzędzia tylko w grupie A: {grupa_a - grupa_b}")
print(f"Narzędzia tylko w grupie A: {grupa_a.difference(grupa_b)}")
print(f"Narzędzia które NIE są jednocześnie w A i B: {grupa_b ^ grupa_a}")
print(f"Narzędzia które NIE są jednocześnie w A i B: {grupa_a.symmetric_difference(grupa_b)}")

print(f"Czy zbiory są rozłączne?: {grupa_a.isdisjoint(grupa_b)}")
```

uzytkownika = {'lr': 0.001, 'batch_size': 64, 'momentum': 0.9}

1. Stwórz finalną konfigurację łącząc słowniki operatorem | (wartości z uzytkownika mają wygrać).  
2. Stwórz drugą wersję finalnej konfiguracji używając .update() (pamiętaj o .copy() , żeby nie zmienić

3. Sprawdź, że obie wersje są równe (operator == ).  
4. Wypisz finalną konfigurację w postaci: klucz: wartość dla wybranych trzech kluczy ( lr ,

```python
domyslne = {'lr': 0.01, 'epochs': 100, 'optimizer': 'sgd', 'batch_size': 32}
uzytkownika = {'lr': 0.001, 'batch_size': 64, 'momentum': 0.9}

finalny = domyslne | uzytkownika
print(finalny)
finalny_v2=domyslne.copy()
finalny_v2.update(uzytkownika)
print(finalny_v2)

print(finalny == finalny_v2)
print(f"Finalna konfiguracja: \n lr: {finalny['lr']}\n batch_size: {finalny['batch_size']}\n momentum: {finalny['momentum']}")
```
