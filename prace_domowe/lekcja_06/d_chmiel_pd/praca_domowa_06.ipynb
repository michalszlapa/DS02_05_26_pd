# `Zadanie 5`

```python
grupa_a = {'python', 'pandas', 'numpy', 'sql', 'tableau'}
grupa_b = {'python', 'r', 'sql', 'ggplot2', 'dplyr'}

suma_operator = grupa_a | grupa_b
wspolne_operator = grupa_a & grupa_b
tylko_a_operator = grupa_a - grupa_b
symetryczna_operator = grupa_a ^ grupa_b

suma_metoda = grupa_a.union(grupa_b)
wspolne_metoda = grupa_a.intersection(grupa_b)
tylko_a_metoda = grupa_a.difference(grupa_b)
symetryczna_metoda = grupa_a.symmetric_difference(grupa_b)

rozlaczne = grupa_a.isdisjoint(grupa_b)

print("Suma zbiorów - operator |:", suma_operator)
print("Suma zbiorów - metoda union():", suma_metoda)

print("Wspólne narzędzia - operator &:", wspolne_operator)
print("Wspólne narzędzia - metoda intersection():", wspolne_metoda)

print("Narzędzia tylko w grupie A - operator -:", tylko_a_operator)
print("Narzędzia tylko w grupie A - metoda difference():", tylko_a_metoda)

print("Narzędzia tylko w jednej z grup - operator ^:", symetryczna_operator)
print("Narzędzia tylko w jednej z grup - metoda symmetric_difference():", symetryczna_metoda)

print("Czy zbiory są rozłączne?:", rozlaczne)
```
