# Zadanie 7  Długości słów

Masz listę słów. Używając list comprehension, stwórz nową listę zawierającą długości tych słów.

Dataset:
`slowa = ["python", "data", "science", "machine", "learning", "AI"]`


Wymagania:
Użyj list comprehension: `[len(slowo) for slowo in slowa]`

Dodatkowo: dict comprehension słowo → długość

Oczekiwany rezultat:
`Lista długości: [6, 4, 7, 7, 8, 2]
Słownik: {"python": 6, "data": 4, ...}`

```python
# Zad 7
slowa = ["python", "data", "science", "machine", "learning", "AI"]
dlugosci_slow = [len(slowo) for slowo in slowa]

dict_slow = {slowo: len(slowo) for slowo in slowa}

print("Lista długości: ", dlugosci_slow)
print("Słownik: ", dict_slow)
```

# Zadanie 8  Map i filter

Używając `map()` i `filter()` z lambda, przetwórz listę liczb: najpierw zostaw tylko dodatnie, potem
podwój każdą.

Dataset:
`liczby = [-5, 3, -2, 8, 0, -1, 7, 4, -9, 6]`

Wymagania:
filter() z lambda: x > 0
map() z lambda: x * 2

Konwertuj wynik na listę

Oczekiwany rezultat:
`Lista: [6, 16, 14, 8, 12]` (tylko dodatnie i podwojone)

```python
# Zad 8
liczby = [-5, 3, -2, 8, 0, -1, 7, 4, -9, 6]
dodatnie = filter(lambda x: x > 0, liczby)
podwojone = map(lambda x: x * 2, dodatnie)

wynik = list(podwojone)

print("Lista: ", wynik)
```

# Zadanie 13  Pipeline przetwarzania danych sprzedażowych

Stwórz kompletny pipeline do analizy danych sprzedażowych. Dane zawierają: dzień, kategoria, liczba transakcji, wartość.

Dataset:
`sprzedaz = [
(1, "elektronika", 45, 12500),
(1, "odziez", 78, 4200),
(2, "elektronika", 52, 15200),
(2, "odziez", 65, 3800),
(3, "elektronika", 38, 9800),
(3, "odziez", 82, 5100),
(4, "elektronika", 61, 18900),
(4, "odziez", 71, 4500),
]`

Wymagania:
- Funkcja `tuple_to_dict()` - konwersja tuple na dict
- Funkcja `dodaj_srednia_transakcji()` - feature engineering (wartosc / liczba_transakcji)
- Funkcja `agreguj_po_kategorii()` - suma wartości per kategoria (użyj dict.get() )

Użyj comprehensions do przetwarzania
Sortowanie kategorii według wartości (malejąco) przez sorted() + lambda

Oczekiwany rezultat:
Ranking kategorii z sumą wartości sprzedaży

```python
# Zad 13
sprzedaz = [
(1, "elektronika", 45, 12500),
(1, "odziez", 78, 4200),
(2, "elektronika", 52, 15200),
(2, "odziez", 65, 3800),
(3, "elektronika", 38, 9800),
(3, "odziez", 82, 5100),
(4, "elektronika", 61, 18900),
(4, "odziez", 71, 4500),
]

dane_przetworzne = [tuple_to_dict(t) for t in sprzedaz]
dane_przetworzne = dodaj_srednia_transakcji(dane_przetworzne)
podsumowanie = agreguj_po_kategorii(dane_przetworzne)

ranking = sorted(podsumowanie.items(), key=lambda x: x[1], reverse=True)

print("Ranking kategorii (kategoria: suma sprzedaży):")
print("-" * 50)
for kategoria, wartosc in ranking:
    print(f"{kategoria}: {wartosc} PLN")




def tuple_to_dict(tup):
    return dict(zip(["dzien", "kategoria", "liczba_transakcji", "wartosc"], tup))


def dodaj_srednia_transakcji(sprzedaz_dict):
    for item in sprzedaz_dict:
        item["srednia_transakcji"] = item["wartosc"] / item["liczba_transakcji"]
    return sprzedaz_dict

def agreguj_po_kategorii(sprzedaz_dict):
    agregacja = {}
    for item in sprzedaz_dict:
        kategoria = item["kategoria"]
        wartosc = item["wartosc"]
        agregacja[kategoria] = agregacja.get(kategoria, 0) + wartosc
    return agregacja
```
