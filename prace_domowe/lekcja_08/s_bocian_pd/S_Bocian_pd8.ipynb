```python
sprzedaz = [
    (1, "elektronika", 45, 12500),
    (1, "odziez", 78, 4200),
    (2, "elektronika", 52, 15200),
    (2, "odziez", 65, 3800),
    (3, "elektronika", 38, 9800),
    (3, "odziez", 82, 5100),
    (4, "elektronika", 61, 18900),
    (4, "odziez", 71, 4500),
]

def tuple_to_dict(dane):
    return [
        {
            "dzien": dzien,
            "kategoria": kategoria,
            "liczba_transakcji": liczba_transakcji,
            "wartosc": wartosc
        }
        for dzien, kategoria, liczba_transakcji, wartosc in dane
    ]

def dodaj_srednia_transakcji(dane):
    return [
        {
            **rekord,
            "srednia_transakcja": round(rekord["wartosc"] / rekord["liczba_transakcji"],2)
        }
        for rekord in dane
    ]

def agreguj_po_kategorii(dane):
    wynik = {}

    for rekord in dane:
        kategoria = rekord["kategoria"]
        wynik[kategoria] = wynik.get(kategoria, 0) + rekord["wartosc"]

    return wynik

dane_dict = tuple_to_dict(sprzedaz)
dane_srednia = dodaj_srednia_transakcji(dane_dict)
agregacja = agreguj_po_kategorii(dane_srednia)

ranking = sorted(
    agregacja.items(),
    key=lambda x: x[1],
    reverse=True
)

print("Dane po konwersji:")
for rekord in dane_dict:
    print(rekord)

print("\nDane ze średnią transakcją:")
for rekord in dane_srednia:
    print(rekord)

print("\nSuma wartości sprzedaży według kategorii:")
print(agregacja)

print("\nRanking kategorii:")
print(ranking)
```
