### Zadanie 7 – Długości słów  
Masz listę słów. Używając list comprehension, stwórz nową listę zawierającą długości tych słów.

Dataset:  
slowa = ["python", "data", "science", "machine", "learning", "AI"]  

Wymagania:Użyj list comprehension: [len(slowo) for slowo in slowa]  
Dodatkowo: dict comprehension słowo → długość  

Oczekiwany rezultat:
Lista długości: [6, 4, 7, 7, 8, 2]  
Słownik: {"python": 6, "data": 4, ...}

```python
slowa = ["python", "data", "science", "machine", "learning", "AI"] 

dlugosci = [len(slowo) for slowo in slowa]
slownik_dlugosci = {slowo: len(slowo) for slowo in slowa}

print(dlugosci)
print(slownik_dlugosci)
```

### Zadanie 8 – Map i filter  
Używając map() i filter() z lambda, przetwórz listę liczb: najpierw zostaw tylko dodatnie, potem  
podwój każdą.

Dataset:  
liczby = [-5, 3, -2, 8, 0, -1, 7, 4, -9, 6]  

Wymagania:  
filter()  
map()  
z lambda: x > 0  
z lambda: x * 2  

Konwertuj wynik na listę  

Oczekiwany rezultat:
Lista: [6, 16, 14, 8, 12] (tylko dodatnie i podwojone)

```python
liczby = [-5, 3, -2, 8, 0, -1, 7, 4, -9, 6]

tylko_dodatnie = filter(lambda x: x > 0, liczby)
dodatnie_i_podwojone = map(lambda x: x * 2, tylko_dodatnie)

wynik = list(dodatnie_i_podwojone)

# print(list(filter(lambda x: x > 0, liczby)))
print(wynik)
```

### Zadanie 13 – Pipeline przetwarzania danych sprzedażowych  
Stwórz kompletny pipeline do analizy danych sprzedażowych. Dane zawierają: dzień, kategoria, liczba  
transakcji, wartość.  

Dataset:  
sprzedaz = [  
(1, "elektronika", 45, 12500),  
(1, "odziez", 78, 4200),  
(2, "elektronika", 52, 15200),  
(2, "odziez", 65, 3800),  
(3, "elektronika", 38, 9800),  
(3, "odziez", 82, 5100),  
(4, "elektronika", 61, 18900),  
(4, "odziez", 71, 4500),  
]

Wymagania:  
Funkcja tuple_to_dict() - konwersja tuple na dict  
Funkcja dodaj_srednia_transakcji() - feature engineering (wartosc / liczba_transakcji)  
Funkcja agreguj_po_kategorii() - suma wartości per kategoria (użyj dict.get() )  
Użyj comprehensions do przetwarzania  
Sortowanie kategorii według wartości (malejąco) przez sorted() + lambda  
Oczekiwany rezultat:Ranking kategorii z sumą wartości sprzedaży

```python
def tuple_to_dict(rekord):
    return {
        "dzien": rekord[0],
        "kategoria": rekord[1],
        "liczba_transakcji": rekord[2],
        "wartosc": rekord[3]
    }

def dodaj_srednia_transakcji(rekord):
    rekord["srednia_transakcja"] = rekord["wartosc"] / rekord["liczba_transakcji"]
    return rekord


def agreguj_po_kategorii(dane):
    agregacja = {}

    for rekord in dane:
        kategoria = rekord["kategoria"]
        wartosc = rekord["wartosc"]

        agregacja[kategoria] = agregacja.get(kategoria, 0) + wartosc

    return agregacja

sprzedaz = [  
(1, "elektronika", 45, 12500),  
(1, "odziez", 78, 4200),  
(2, "elektronika", 52, 15200),  
(2, "odziez", 65, 3800),  
(3, "elektronika", 38, 9800),  
(3, "odziez", 82, 5100),  
(4, "elektronika", 61, 18900),  
(4, "odziez", 71, 4500),  
]


# 1. Konwersja tuple -> dict (lista słowników)
sprzedaz_dict = [tuple_to_dict(rekord) for rekord in sprzedaz]

# 2. Dodanie średniej wartości transakcji, rekordem jest słownik, do którego dodajemy klucz: średnia wartość i: wartość
sprzedaz_z_srednia = [dodaj_srednia_transakcji(rekord) for rekord in sprzedaz_dict]

# 3. Agregacja po kategorii
suma_po_kategorii = agreguj_po_kategorii(sprzedaz_z_srednia)

# 4. Sortowanie malejąco po wartości
ranking = sorted(
    suma_po_kategorii.items(),
    key=lambda x: x[1],
    reverse=True
)

print("Ranking kategorii:")

for kategoria, suma in ranking:
    print(kategoria, suma)
```
