```python
#Zadanie 8 ^ Map i filter
liczby = [-5, 3, -2, 8, 0, -1, 7, 4, -9, 6]

dodatnie = filter(lambda x: x > 0, liczby)

podwojone = map(lambda x: x * 2, dodatnie)

wynik = list(podwojone)

print("Lista:", wynik)
```

```python
#Zadanie 13 ^ Pipeline przetwarzania danych sprzedażowych
sprzedaz = [
    (1, "elektronika", 45, 12500),
    (1, "odziez", 78, 4200),
    (2, "elektronika", 52, 15200),
    (2, "odziez", 65, 3800),
    (3, "elektronika", 38, 9800),
    (3, "odziez", 82, 5100),
    (4, "elektronika", 61, 18900),
    (4, "odziez", 71, 4500),
]


def tuple_to_dict(rekord):
    dzien = rekord[0]
    kategoria = rekord[1]
    liczba_transakcji = rekord[2]
    wartosc = rekord[3]

    return {
        "dzien": dzien,
        "kategoria": kategoria,
        "liczba_transakcji": liczba_transakcji,
        "wartosc": wartosc
    }


def dodaj_srednia_transakcji(rekord):
    rekord["srednia_transakcji"] = rekord["wartosc"] / rekord["liczba_transakcji"]
    return rekord


def agreguj_po_kategorii(dane):
    wynik = {}

    for rekord in dane:
        kategoria = rekord["kategoria"]
        wartosc = rekord["wartosc"]

        if kategoria not in wynik:
            wynik[kategoria] = 0

        wynik[kategoria] = wynik[kategoria] + wartosc

    return wynik


dane = []

for rekord in sprzedaz:
    nowy_rekord = tuple_to_dict(rekord)
    nowy_rekord = dodaj_srednia_transakcji(nowy_rekord)
    dane.append(nowy_rekord)


suma_kategorii = agreguj_po_kategorii(dane)

ranking = sorted(
    suma_kategorii.items(),
    key=lambda x: x[1],
    reverse=True
)

print("Ranking kategorii:")

for kategoria, suma in ranking:
    print(kategoria, "-", suma, "zł")
```
