***Zadanie 7***

Dla listy słów stwórz listę ich długości oraz słownik słowo → długość.\
Dataset: ["python", "data", "science", "machine", "learning", "AI"]\
Oczekiwany rezultat: [6, 4, 7, 7, 8, 2] i {"python": 6, "data": 4, ...}

```python
from itertools import product

dataset_7 = ["python", "data", "science", "machine", "learning", "AI"]

dlugosc_7 = [len(x) for x in dataset_7]

slownik_7 = {data: dlug for data,dlug in zip(dataset_7,dlugosc_7)}

print(f'Spis długości: {dlugosc_7}')
print(f'Slownik: {slownik_7}')
```

***Zadanie 8***

Z listy liczb zostaw tylko dodatnie, a następnie podwój każdą z nich.\
Dataset: [-5, 3, -2, 8, 0, -1, 7, 4, -9, 6]\
Oczekiwany rezultat: [6, 16, 14, 8, 12]

```python
dataset_8 = [-5, 3, -2, 8, 0, -1, 7, 4, -9, 6]

answer_8 = [x*2 for x in dataset_8 if x>0]

print(answer_8)
```
