Zadanie 7  Długości słów

```python
slowa = ["python", "data", "science", "machine", "learning", "AI"]
dlugosci = [len(slowo) for slowo in slowa]
dlugosci2 = []
for i in slowa:
    dlugosci2.append(len(i))
print(f"List comprehension : {dlugosci}")
print(f"Porównanie normalna petla : {dlugosci2}")

#dictionary comprehension
dict_comp = {slowa: dlugosci for slowa, dlugosci in zip(slowa, dlugosci)}
for slowa, dlugosci in dict_comp.items():
  print(f"{slowa} : {dlugosci}")
```

```python
liczby = [-5, 3, -2, 8, 0, -1, 7, 4, -9, 6]
wynik = list(
    map(lambda x: x * 2,
        filter(lambda x: x > 0, liczby)))

print(wynik)
```

Zadanie 13  Pipeline przetwarzania danych sprzedażowych

```python
sprzedaz = [
    (1, "elektronika", 45, 12500),
    (1, "odziez", 78, 4200),
    (2, "elektronika", 52, 15200),
    (2, "odziez", 65, 3800),
    (3, "elektronika", 38, 9800),
    (3, "odziez", 82, 5100),
    (4, "elektronika", 61, 18900),
    (4, "odziez", 71, 4500),
]
def tuple_to_dict(rekord):
    return {
        "dzien": rekord[0],
        "kategoria": rekord[1],
        "liczba_transakcji": rekord[2],
        "wartosc": rekord[3]
    }
def dodaj_srednia_transakcji(rekord):
    rekord["srednia_transakcji"] = (
        rekord["wartosc"] / rekord["liczba_transakcji"]
    )
    return rekord

dane = [tuple_to_dict(r) for r in sprzedaz]
dane = [dodaj_srednia_transakcji(r) for r in dane]

def agreguj_po_kategorii(dane):
    wynik = {}

    for rekord in dane:
        kategoria = rekord["kategoria"]

        wynik[kategoria] = (
            wynik.get(kategoria, 0)
            + rekord["wartosc"]
        )

    return wynik

agregacja = agreguj_po_kategorii(dane)

ranking = sorted(
    agregacja.items(),
    key=lambda x: x[1],
    reverse=True
)

print("Ranking kategorii:")

for kategoria, wartosc in ranking:
    print(f"{kategoria}: {wartosc}")
```
