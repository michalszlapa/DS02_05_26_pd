# Zadanie 1 – Kalkulator BMI
### Napisz program, który obliczy wskaźnik BMI (Body Mass Index).

Dane:  
-> Waga: 70 kg  
-> Wzrost: 1.75 m  

Wymagania:  
- Stwórz zmienne waga i wzrost  
- Oblicz BMI według wzoru: BMI = waga / (wzrost²)  
- Wyświetl wynik z 2 miejscami po przecinku  
- Dodaj interpretację: BMI < 18.5 (niedowaga), 18.5-24.9 (norma), ≥25 (nadwaga)

```python
waga = 70
wzrost=1.75

bmi=waga/(wzrost**2)
print(round(bmi, 2))

# Zapis bmi z przedziału >= 18.5 i < 25 można uprościć : bmi >= 18.5 jeśli wcześniej mamy if/elif dotyczący bmi >= 25
if bmi >= 25:
    interpretacja = "nadwaga"
elif bmi >= 18.5:
    interpretacja = "norma"
else:
    bmi < 18.5
    interpretacja = "niedowaga"

print(interpretacja)
```

# Zadanie 7 – Porównanie dwóch modeli
### Określ, który model jest lepszy.

Dane:  
Model A: accuracy = 0.87, train time = 120s  
Model B: accuracy = 0.89, train time = 450s  

Wymagania:  
- Porównaj accuracy (różnica > 0.02 jest znacząca)  
- Porównaj czas treningu  
- Użyj operatorów logicznych do określenia "najlepszego kompromisu"  
- Wyświetl rekomendację

```python
accuracy_a = 0.87
accuracy_b = 0.89
train_time_a = 120
train_time_b = 450

roznica = round(accuracy_b - accuracy_a, 4)

if roznica > 0.02:
    print("Model B ma znacząco lepszą dokładność.")
elif roznica < -0.02:
    print("Model A ma znacząco lepszą dokładność.")
else:
    print("Różnica dokładności nie jest znacząca.")

if train_time_a < train_time_b:
    print("Model A trenuje się szybciej.")
elif train_time_b < train_time_a:
    print("Model B trenuje się szybciej.")
else:
    print("Oba modele mają taki sam czas treningu.")

if roznica <= 0.02 and train_time_a < train_time_b:
    print("Rekomendacja: Model A, ponieważ różnica accuracy nie jest znacząca, a czas treningu jest dużo krótszy.")
elif roznica > 0.02:
    print("Rekomendacja: Model B, ponieważ ma znacząco lepszą accuracy.")
else:
    print("Rekomendacja: potrzebna dodatkowa analiza.")


```

# Zadanie 8 – Obliczanie learning rate
### Oblicz learning rate dla schedulera.

Dane:  
- Początkowy learning rate: 0.1  
- Współczynnik decay: 0.95  
- Numer epoki: 10  

Wymagania:  
-> Użyj wzoru: lr_new = lr_initial × (decay ^ epoch)  
-> Operator potęgowania **  
-> Wyświetl wynik w notacji naukowej ( .2e )  

```python
lr_initial = 0.1
decay = 0.95
epoch = 10

lr_new = lr_initial * (decay ** epoch)

print(f"Nowa wartość learning rate po {epoch} epokach: {lr_new:.2e}")
```

# Zadanie 9 – Normalizacja Min-Max
### Znormalizuj wartość do zakresu [0, 1].

Dane:  
-> Wartość: 75  
-> Minimum w kolumnie: 50  
-> Maximum w kolumnie: 100  

Wymagania:  
- Wzór: normalized = (value - min) / (max - min)  
- Oblicz dla wartości 75, 50, 100, 62.5  
- Sprawdź, czy wynik mieści się w [0, 1]  
- Wyświetl tabelkę: wartość oryginalna | wartość znormalizowana

```python
wartosci = [75, 50, 100, 62.5]
minimum = 50
maksimum = 100

print(f"{'Oryginalna':<12} {'Znormalizowana':<16} {'Status'}")

for wartosc in wartosci:
    normalized = (wartosc - minimum) / (maksimum - minimum)

    if 0 <= normalized <= 1:
        status = "mieści się"
    else:
        status = "nie mieści się"

    print(f"{wartosc:<12} {normalized:<16.2f} {status}")
```

# Zadanie 11 – Sprawdzanie poprawności hiperparametrów
### Zwaliduj hiperparametry przed treningiem.

Dane:  
learning_rate = 0.01  
batch_size = 64  
epochs = 100  
dropout = 0.5  

Wymagania:  
-> Sprawdź czy learning_rate w zakresie (0, 1]  
-> Sprawdź czy batch_size jest potęgą 2 (użyj operacji bitowych lub logarytmów)  
-> Sprawdź czy epochs > 0 i < 1000  
-> Sprawdź czy dropout w zakresie [0, 1]  
-> Wyświetl "OK" lub komunikat błędu dla każdego parametru

```python
learning_rate = 0.01
batch_size = 64
epochs = 100
dropout = 0.5

if 0 < learning_rate <= 1:
    print("learning_rate: OK")
else:
    print("learning_rate: błąd - wartość musi być w zakresie (0, 1]")

if batch_size > 0 and batch_size & (batch_size - 1) == 0:
    print("batch_size: OK")
else:
    print("batch_size: błąd - wartość musi być potęgą 2")

if 0 < epochs < 1000:
    print("epochs: OK")
else:
    print("epochs: błąd - wartość musi być większa od 0 i mniejsza od 1000")

if 0 <= dropout <= 1:
    print("dropout: OK")
else:
    print("dropout: błąd - wartość musi być w zakresie [0, 1]")
```

# Zadanie 12 – Estymacja czasu treningu
### Oblicz szacowany czas treningu modelu.

Dane:  
- Liczba próbek: 50000  
- Batch size: 128  
- Liczba epok: 50  
- Czas na jeden batch: 0.15 sekundy  

Wymagania:  
-> Oblicz liczbę batchy na epokę (użyj dzielenia całkowitego i modulo)  
-> Oblicz całkowity czas w sekundach  
-> Skonwertuj na minuty i godziny  
-> Wyświetl: "Szacowany czas: X h Y min Z s"

```python
liczba_probek = 50000
batch_size = 128
liczba_epok = 50
batch_time = 0.15

pelne_batche = liczba_probek // batch_size
reszta = liczba_probek % batch_size

if reszta > 0:
    batche_na_epoke = pelne_batche + 1
else:
    batche_na_epoke = pelne_batche

wszystkie_batche = batche_na_epoke * liczba_epok
czas_sekundy = wszystkie_batche * batch_time

godziny = int(czas_sekundy // 3600)
pozostale_sekundy = czas_sekundy % 3600

minuty = int(pozostale_sekundy // 60)
sekundy = pozostale_sekundy % 60

print("Liczba pełnych batchy na epokę:", pelne_batche)
print("Reszta próbek:", reszta)
print("Liczba batchy na epokę:", batche_na_epoke)
print("Całkowity czas w sekundach:", czas_sekundy)

print(f"Szacowany czas: {godziny} h {minuty} min {sekundy:.1f} s")

```
