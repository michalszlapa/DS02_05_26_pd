### Zadanie 7 – Klasa Pracownik  
Stwórz klasę Pracownik z atrybutami: imie , stanowisko , pensja .  
Metody:  
podwyzka(procent)  
__str__()  
– zwiększa pensję o procent%  
– reprezentacja tekstowa  
Stwórz 3 pracowników i daj im podwyżki.  

```python
class Pracownik:

    def __init__(self, imie: str, stanowisko: str, pensja: float):
        self.imie = imie
        self.stanowisko = stanowisko
        self.pensja = pensja

    def podwyzka(self, procent: float):
        self.pensja += self.pensja * (procent/100)

    def __str__(self):
        return f"Pracownik: {self.imie} | Stanowisko: {self.stanowisko}  | Pensja: {self.pensja:.2f} zł"
    
# 1. Tworzenie 3 obiektów pracowników
pracownik1 = Pracownik("Kacper", "Programista", 8000)
pracownik2 = Pracownik("Monika", "Manager", 12000)
pracownik3 = Pracownik("Kuba", "Tester", 6000)

print("Stan początkowy:")
print(pracownik1)
print(pracownik2)
print(pracownik3)

print("\n--- Przyznawanie podwyżek ---")

# 2. Dajemy pracownikom podwyżki (np. 10%, 15% i 5%)
pracownik1.podwyzka(10)
pracownik2.podwyzka(15)
pracownik3.podwyzka(5)

print("\nStan po podwyżkach:")
print(pracownik1)
print(pracownik2)
print(pracownik3)   
```
