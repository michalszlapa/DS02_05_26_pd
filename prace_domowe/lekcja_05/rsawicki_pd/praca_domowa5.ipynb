# Zadanie 2 – Dodawanie elementów do listy
### Zacznij od pustej listy results = [] .  
### Dodaj do niej 5 wartości accuracy: 0.75, 0.82, 0.79, 0.85, 0.88 (użyj append ). 

Następnie wyświetl:  
- długość listy ( len ),  
- średnią wartość ( sum / len ),  
- maksymalną wartość ( max ),  
- minimalną wartość ( min ).

```python
results = []
results.append(0.75)
results.append(0.82)
results.append(0.79)
results.append(0.85)
results.append(0.88)

srednia = sum(results) / len(results)

print("Długość listy:", len(results))
print("Średnia wartość:", srednia)
print("Maksymalna wartość:", max(results))
print("Minimalna wartość:", min(results))
```

# Zadanie 9 – Confusion matrix i metryki
Masz confusion matrix dla klasyfikacji binarnej:

cm = [  
    [80, 10],  #TN, FP  
    [5, 45]  #FN, TP  
]

Wyciągnij wartości TN, FP, FN, TP przez dwuwymiarowe indeksowanie i oblicz:
- accuracy = (TP + TN) / (TP + TN + FP + FN) ,
- precision = TP / (TP + FP) ,recall = TP / (TP + FN)

```python
cm = [
    [80, 10],  # TN, FP
    [5, 45]    # FN, TP
]

TN = cm[0][0]
FP = cm[0][1]
FN = cm[1][0]
TP = cm[1][1]

accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)

print("TN:", TN)
print("FP:", FP)
print("FN:", FN)
print("TP:", TP)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```

# Zadanie 10 – Łączenie i dzielenie listy
Masz listę nazw modeli: ['LogisticRegression', 'RandomForest', 'XGBoost', 'SVM'].  

Zbuduj:

- string "LogisticRegression, RandomForest, XGBoost, SVM" (użyj join ),  
- string SQL: "SELECT model FROM (LogisticRegression, RandomForest, XGBoost, SVM)" — podstaw
złączoną listę w odpowiednie miejsce przez f-string.

Następnie podziel wynikowy string przez split(', ') i sprawdź, czy otrzymałaś/eś z powrotem
identyczną listę (porównaj operatorem == — otrzymasz True / False , które wyświetl).

```python
modele = ['LogisticRegression', 'RandomForest', 'XGBoost', 'SVM']

polaczone_modele = ", ".join(modele)

sql = f"SELECT model FROM ({polaczone_modele})"

lista_po_split = polaczone_modele.split(", ")

czy_identyczna = lista_po_split == modele

print(polaczone_modele)
print(sql)
print(lista_po_split)
print(czy_identyczna)
```

# Zadanie 11 – Parsowanie linii logu
Masz string logu: "2024-01-15 10:23:45 ERROR Training failed due to OOM" .   

-> Użyj split(' ', 3) (z maxsplit=3 ), aby uzyskać listę 4-elementową: [data, czas, poziom, wiadomość] .   
-> Wyciągnij poszczególne pola i wyświetl w formacie:

[2024-01-15 10:23:45] ERROR: Training failed due to OOM

Dodatkowo sprawdź, czy poziom jest jednym z typowych wartości — użyj operatora in z listą ['INFO',
'WARNING', 'ERROR', 'DEBUG'] i zapisz wynik ( True / False ) do zmiennej poziom_znany , którą
wyświetlisz.

```python
log = "2024-01-15 10:23:45 ERROR Training failed due to OOM"

data, czas, poziom, wiadomosc = log.split(" ", 3)

print(f"[{data} {czas}] {poziom}: {wiadomosc}")

typowe_poziomy = ['INFO', 'WARNING', 'ERROR', 'DEBUG']
poziom_znany = poziom in typowe_poziomy

print(poziom_znany)
```

# Zadanie 12 – Budowanie zapytania SQL

Masz tabelę "customers" , listę kolumn ['id', 'name', 'email', 'city'] oraz warunek "city = 'New
York'" . 

Użyj ', '.join(kolumny) i f-stringa, aby zbudować zapytanie:  
- SELECT id, name, email, city FROM customers WHERE city = 'New York'

```python
tabela = "customers"
kolumny = ['id', 'name', 'email', 'city']
warunek = "city = 'New York'"

kolumny_sql = ", ".join(kolumny)

zapytanie = f"SELECT {kolumny_sql} FROM {tabela} WHERE {warunek}"

print(zapytanie)
```
