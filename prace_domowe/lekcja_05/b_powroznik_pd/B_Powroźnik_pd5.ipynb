Zadanie 7. Zliczanie wystąpień
Masz listę predykcji: ['spam', 'ham', 'spam', 'spam', 'ham', 'ham', 'spam'] . Użyj metody count() ,
aby policzyć wystąpienia 'spam' i 'ham' . Wyświetl wynik jako: "spam: X, ham: Y" .

Zadanie 9  Confusion matrix i metryki
Masz confusion matrix dla klasyfikacji binarnej:
cm = [
[80, 10], # TN, FP
[5, 45] # FN, TP
]
Wyciągnij wartości TN, FP, FN, TP przez dwuwymiarowe indeksowanie i oblicz:
accuracy = (TP + TN) / (TP + TN + FP + FN) ,
precision = TP / (TP + FP) ,
recall = TP / (TP + FN) .
Wyświetl wyniki w formacie procentowym z dwoma miejscami po przecinku.

```python
cm = [
[80, 10], # TN, FP
[5, 45] # FN, TP
]

tn = cm[0][0]
fp = cm[0][1]
fn = cm[1][0]
tp = cm[1][1]

print(f"TN = {tn}, FP = {fp}, FN = {fn}, TP = {tp}")
accuracy = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp)
recall = tp / (tp + fn)

print(f"Accuracy : {accuracy:.2%}")
print(f"Precision: {precision:.2%}")
print(f"Recall : {recall:.2%}")

```
