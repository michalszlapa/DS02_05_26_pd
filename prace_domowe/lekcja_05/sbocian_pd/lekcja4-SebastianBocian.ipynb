```python
#zadanie 20
table = "patients"

columns = ["patient_id", "age", "diagnosis"]

condition1 = "age > 40"

diagnoses = ["diabetes", "hypertension"]
condition2 = f"diagnosis IN ({', '.join([repr(d) for d in diagnoses])})"

order_by = "age DESC"

limit = 100

query = f"""
SELECT {', '.join(columns)}
FROM {table}
WHERE {condition1}
AND {condition2}
ORDER BY {order_by}
LIMIT {limit};
"""

print(query)
```

```python
#zadanie 12 
samples = 50000
batch_size = 128
epochs = 50
time_per_batch = 0.15

# liczba batchy na epokę
batches = samples // batch_size

if samples % batch_size != 0:
    batches += 1

# całkowity czas
total_seconds = batches * epochs * time_per_batch

# konwersja
hours = int(total_seconds // 3600)
minutes = int((total_seconds % 3600) // 60)
seconds = int(total_seconds % 60)

print(f"Szacowany czas: {hours} h {minutes} min {seconds} s")
```
