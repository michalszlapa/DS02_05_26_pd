## Zadanie 9 – Confusion matrix i metryki

Masz confusion matrix dla klasyfikacji binarnej:

cm
=
[
[
80
,
10
]
,
# TN, FP
[
5

# FN, TP
]
* Wyciągnij wartości TN, FP, FN, TP przez dwuwymiarowe indeksowanie i oblicz:
  accuracy = (TP + TN) / (TP + TN + FP + FN)

  precision = TP / (TP + FP)
  
  recall = TP / (TP + FN)

* Wyświetl wyniki w formacie procentowym z dwoma miejscami po przecinku.
(średnie)

```python

cm = [
    [80, 10],  # TN, FP
    [5, 45]    # FN, TP
]

TN = cm[0][0]
FP = cm[0][1]
FN = cm[1][0]
TP = cm[1][1]

#Metryki 

accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)

#Wyniki

print(f"Accuracy: {accuracy*100:.2f}%")
print(f"Precision: {precision*100:.2f}%")
print(f"Recall: {recall*100:.2f}%")
```

## Zadanie 14 – Ekstrakcja metadanych z nazwy pliku
Masz ścieżkę:
"/data/processed/2024/experiment_42_v3.csv"
# Za pomocą split('/'), indeksowania imetod stringów wyciągnij:

* nazwę pliku (
"experiment_42_v3.csv")

* katalog ("/data/processed/2024")

* nazwę bez rozszerzenia ("experiment_42_v3")

* rozszerzenie (".csv")

* rok z katalogu ("2024").

* Wyświetl wszystko w czytelnym raporcie (f-string z wyrównaniem).

```python

sciezka = "/data/processed/2024/experiment_42_v3.csv"

# Podział ścieżki
czesci = sciezka.split('/')

plik = czesci[-1]
rok = czesci[-2]
katalog = '/'.join(czesci[:-1])


nazwa_bez_rozszerzenia = plik.split('.')[0]
rozszerzenie = '.' + plik.split('.')[1]


print(f"{'Nazwa pliku:':25} {plik}")
print(f"{'Katalog:':25} {katalog}")
print(f"{'Nazwa bez rozszerzenia:':25} {nazwa_bez_rozszerzenia}")
print(f"{'Rozszerzenie:':25} {rozszerzenie}")
print(f"{'Rok z katalogu:':25} {rok}")
```
