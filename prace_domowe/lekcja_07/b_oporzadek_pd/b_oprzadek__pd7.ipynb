```python
lista = [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3]
suma = 0
ilosc = 0
for liczba in lista:
  if liczba > 0:
    suma = suma + liczba
    ilosc += 1
print(suma)
print(ilosc)
```

ZADANIE 5


```python
n=50
suma=0
i=0
while i <= n:
  suma += i
  i += 1

wzor = n*(n+1)/2
print(f"Suma pętli : {suma}")
print(f"Suma wzoru : {wzor}")
```
