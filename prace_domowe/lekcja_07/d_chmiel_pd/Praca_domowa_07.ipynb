# `Zadanie 2`

```python
lista = [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3]

suma = 0
ilosc = 0

for temp in lista:
    if temp > 0:
        suma += temp
        ilosc += 1

print("Suma dodatnich wartości:", suma)
print("Liczba dodatnich wartości:", ilosc)
```

# `Zadanie 5`

```python
i = 1
suma = 0

while i <= 50:
    suma += i
    i += 1

print("Suma liczb od 1 do 50:", suma)

sprawdzenie = 50 * (50 + 1) // 2
print("Sprawdzenie wzorem:", sprawdzenie)
```

# `Zadanie 12`

```python
pesele = ['92041812345', '8501015432', 'abc123', '78121698765']

for pesel in pesele:
    if len(pesel) == 11 and pesel.isdigit():
        print(f"PESEL {pesel}: poprawny format")
    else:
        if len(pesel) != 11:
            print(f"PESEL {pesel}: błędny format (długość {len(pesel)})")
        elif not pesel.isdigit():
            print(f"PESEL {pesel}: błędny format (zawiera znaki inne niż cyfry)")
```
