### Zadanie 2. Suma liczb dodatnich
Dla listy [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3]   
oblicz sumę dodatnich wartości (pętla for , if temp >
0 ) oraz ile ich jest. Wypisz oba wyniki.

```python
lista = [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3]
suma = 0
liczba_dodatnich = 0

for liczba in lista:
    if liczba > 0:
        suma += liczba
        liczba_dodatnich += 1

print(f"Suma liczb dodatnich: {suma}")
print(f"Dodatnich liczb jest: {liczba_dodatnich}")
```

### Zadanie 5. Suma przez while  
Używając pętli while (nie for !) oblicz sumę liczb od 1 do 50. Sprawdź wynik wzorem n*(n+1)/2 =
1275 .

```python
suma_liczb = 0
liczba = 1

while liczba <= 50:
    suma_liczb += liczba
    liczba += 1

n = 50
suma_wzor = int(n*(n+1)/2)

czy_obliczenia_poprawne = suma_liczb == suma_wzor

# Wyświetlenie wyników i weryfikacja
print(f"Suma z pętli: {suma_liczb}")
print(f"Suma ze wzoru: {suma_wzor}")
print(f"Czy obliczenia są poprawne? {czy_obliczenia_poprawne}")
```

### Zadanie 12. Walidacja numeru PESEL (uproszczona)  
Dla listy PESEL-i ['92041812345', '8501015432', 'abc123', '78121698765'] :  
Dla każdego numeru sprawdź:  
długość dokładnie 11 znaków,  
wszystkie znaki cyfrowe ( s.isdigit() – metoda z L5).  
Wypisz raport: PESEL 92041812345: poprawny format albo PESEL 8501015432: błędny format (długość 10) .  
Oczekiwany wynik: tylko pierwszy i ostatni są poprawne.

```python
lista_peseli = ['92041812345', '8501015432', 'abc123', '78121698765']

for pesel in lista_peseli:

    if len(pesel) == 11 and pesel.isdigit(): 
        print(f"PESEL {pesel}: poprawny format")
    else: 
        if len(pesel) != 11:
            print(f"PESEL {pesel}: błędny format (długość {len(pesel)})")
        elif not pesel.isdigit():
            print(f"PESEL {pesel}: błędny format (zawiera znaki inne niż cyfry)")
```
