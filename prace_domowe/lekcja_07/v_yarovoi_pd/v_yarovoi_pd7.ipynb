***Zadanie 5. Suma przez while***

Używając pętli while (nie for !) oblicz sumę liczb od 1 do 50. Sprawdź wynik wzorem n*(n+1)/2 =
1275

```python
suma5 = 0
i = 1
n = 50

while i <= n:
    suma5 += i
    i = i+1

weryfikacja = n*(n+1)/2

if weryfikacja == suma5:
    print(f'Zweryfikowano, suma = {suma5}')
else:
    print('ajajaj')
```

***Zadanie 12. Walidacja numeru PESEL (uproszczona)***

Dla listy PESEL-i ['92041812345', '8501015432', 'abc123', '78121698765'] :\
Dla każdego numeru sprawdź:\
długość dokładnie 11 znaków,\
wszystkie znaki cyfrowe ( s.isdigit() – metoda z L5.\
Wypisz raport: PESEL 92041812345: poprawny format albo PESEL 8501015432: błędny format (długość
10) .\
Oczekiwany wynik: tylko pierwszy i ostatni są poprawne.\

```python
pesele = ['92041812345', '8501015432', 'abc123abc12', '78121698765']

#wariant 1: najperw wypisuję wszystkie błędy, a w else daję poprawny format, wydaję się, że sposób jest błędo-twórczy, gdy nie znam wszystkich błędów

for pesel in pesele:
    if not len(pesel) == 11:
        print(f'PESEL {pesel}: błędny format (długość {len(pesel)})')
    else:
        if not pesel.isdigit():
            print(f'PESEL {pesel}: błędny format (są inne znaki, niż liczby)')
        else:
            print(f'PESEL {pesel}: poprawny format')


#wariant 2: najpierw określam pozytywny warunek, a później uzupełniam błędy. Sposób 100% prawidłowy, gdy znam warunki pozytywu, a później można uzupełniać błędy w nieskończoność elifami

print(f'''{'='*30}\nWariant 2\n{'='*30}''')

for pesel in pesele:
    if len(pesel) == 11 and pesel.isdigit():
        print(f'PESEL {pesel}: poprawny format')
    else:
        if not pesel.isdigit():
            print(f'PESEL {pesel}: błędny format (są inne znaki, niż liczby)')
        elif len(pesel) != 11:
            print(f'PESEL {pesel}: błędny format (długość {len(pesel)})')
        else:
            print(f'PESEL {pesel}: błędny format (inne)')
```

***Zadanie 13. Klasyfikacja metodą k najbliższych sąsiadów 1D, k=3***

Dane:\
trening = [(1.5, 'A'), (2.3, 'A'), (3.8, 'B'), (4.2, 'B'), (5.1, 'B')]\
x_nowy = 3.5\
Kroki:\
Zbuduj list comprehension z parami (odleglosc, klasa) : [(abs(x_nowy - x), klasa) for x, klasa
in trening] .\
Posortuj wynik sorted(...) – rosnąco po pierwszym polu (krotki sortują się leksykograficznie).\
Weź k = 3 pierwszych elementów – to najbliżsi sąsiedzi.\
Policz wystąpienia klas w sąsiadach (dwie pętle if lub dwa list comprehensions z sum ).\
Wypisz: 3 najbliższe punkty z odległościami oraz przewidywaną klasę (tę z większą liczbą wystąpień).\

```python
trening = [(1.5, 'A'), (2.3, 'A'), (3.8, 'B'), (4.2, 'B'), (5.1, 'B')]
x_nowy = 3.5

nowy_trening = [(round(abs(x_nowy-x),2),klasa) for x,klasa in trening]
nowy_trening_sorted = sorted(nowy_trening)

k_3 = nowy_trening_sorted[0:3]
a_sum = 0
b_sum = 0

for i in k_3:
    if 'A' in i:
        a_sum += 1
    elif 'B' in i:
        b_sum +=1

a_sum2 = 0
b_sum2 = 0

a_sum2 = sum([1 for i in k_3 if 'A' in i])
b_sum2 = sum([1 for i in k_3 if 'B' in i])

print(f'List comprehensiion: {a_sum2},{b_sum2}')

print(f'''Najbliższe punkty: {k_3}
Przewidywana klasa: {'A' if a_sum>b_sum else 'B'}
''')
```

***Zadanie 14. Korelacja Pearsona ręcznie***

Dane:\
wzrost = [165, 170, 175, 180, 185]\
waga = [60, 65, 70, 75, 80]\
Oblicz współczynnik korelacji Pearsona:\
Oblicz średnie: sr_x = sum(wzrost) / len(wzrost) , sr_y = sum(waga) / len(waga) .\
W jednej pętli for i in range(len(wzrost)) policz trzy sumy: sum_xy , sum_x2 , sum_y2 :\
sum_xy += (wzrost[i] - sr_x) * (waga[i] - sr_y)\
sum_x2 += (wzrost[i] - sr_x) ** 2\
sum_y2 += (waga[i] - sr_y) ** 2\
 r = sum_xy / (sum_x2 * sum_y2) ** 0.5 .\
Wypisz r z dokładnością do 4 miejsc. Oczekiwany wynik: r = 1.0000 (idealna korelacja dodatnia).\

```python
wzrost = [165, 170, 175, 180, 185]
waga = [60, 65, 70, 75, 80]

avg_wzrost = sum(wzrost)/len(wzrost)
avg_waga = sum(waga)/len(waga)

sum_xy = 0
sum_x2 = 0
sum_y2 = 0

for i in range(len(wzrost)):
    sum_xy += (wzrost[i]-avg_wzrost)*(waga[i]-avg_waga)
    sum_x2 += (wzrost[i]-avg_wzrost)**2
    sum_y2 += (waga[i]-avg_waga)**2

r = round(sum_xy / (sum_x2 * sum_y2) ** 0.5)

print(f'r = {r:.4f}. Idealna korelacja dodatnia')
```
