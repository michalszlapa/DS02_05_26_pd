# Zadanie 2. Suma liczb dodatnich
Dla listy `[-5, 3, -2, 8, 12, -1, 6, 0, 15, -3]` oblicz sumę dodatnich wartości (pętla for , if temp > 0 ) oraz ile ich jest. Wypisz oba wyniki.

```python
temp = [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3]

l_temp_dodatnie = 0
sum_temp_dodatnie = 0

for i in temp:
    if i > 0:
        l_temp_dodatnie += 1
        sum_temp_dodatnie += i
print(f'Liczba temperatur dodatnich: {l_temp_dodatnie}, suma temperatur dodatnich: {sum_temp_dodatnie}')
```

# Zadanie 5. Suma przez while
Używając pętli `while` (nie for !) oblicz sumę liczb od 1 do 50. Sprawdź wynik wzorem `n*(n+1)/2 = 1275`

```python
i = 1
n = 50

suma = 0
while i <= n:
    suma += i
    i += 1

print(f'Suma liczb od 1 do {n} wynosi: {suma}')
print(f'Suma używając wzoru n*(n+1)/2 = {n*(n+1)/2}')
```

# Zadanie 12. Walidacja numeru PESEL (uproszczona)
Dla listy PESEL-i `['92041812345', '8501015432', 'abc123', '78121698765']` :
Dla każdego numeru sprawdź:
długość dokładnie 11 znaków,
wszystkie znaki cyfrowe ( `s.isdigit()` – metoda z L5.
Wypisz raport: PESEL 92041812345: poprawny format albo PESEL 8501015432: błędny format (długość
10) .
Oczekiwany wynik: tylko pierwszy i ostatni są poprawne

```python
pesel = ['92041812345', '8501015432', 'abc123', '78121698765']

for p in pesel:
    message = "poprawny format" if len(p) == 11 and p.isdigit() else f'błędny format (długość {len(p)})'
    print(f'PESEL {p}: {message}')
```
