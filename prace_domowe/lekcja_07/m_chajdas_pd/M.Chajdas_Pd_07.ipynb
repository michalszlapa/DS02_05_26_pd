## Zadanie 2. Suma liczb dodatnich
# Dla listy [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3] :

* oblicz sumę dodatnich wartości (pętla for,if temp >0)
* oraz ile ich jest.
*  Wypisz oba wyniki.

```python
dane = [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3]

suma_dodatnich = 0
liczba_dodatnich = 0

for temp in dane:
    if temp > 0:
        suma_dodatnich += temp
        liczba_dodatnich += 1

print(*f"Suma dodatnich=", suma_dodatnich)
print(*f"Liczba dodatnich=", liczba_dodatnich)
```

## Zadanie 5. Suma przez while

#Używając pętli while (nie for!):
* oblicz sumę liczb od 1 do 50.
*  Sprawdź wynik wzorem
   n*(n+1)/2 =1275


```python
suma = 0
i = 1

while i <= 50:
    suma += i
    i += 1

print(*f"Suma =", suma)
```

```python
wzor = 50 * (50 + 1) / 2
print(*f"Wynik ze wzoru =", wzor)
```

## Zadanie 12. Walidacja numeru PESEL (uproszczona)
*  Dla listy PESEL-i ['92041812345', '8501015432', 'abc123', '78121698765'] :
# Dla każdego numeru sprawdź:
* długość dokładnie 11 znaków,wszystkie znaki cyfrowe (s.isdigit() – metoda z L5).

# Wypisz raport:
- PESEL 92041812345: poprawny format
  albo
- PESEL 8501015432: błędny format (długość10)

* Oczekiwany wynik: tylko pierwszy i ostatni są poprawne.

```python
pesele = ['92041812345', '8501015432', 'abc123', '78121698765']

for pesel in pesele:
    if len(pesel) == 11 and pesel.isdigit():
        print(f"PESEL {pesel}: poprawny format")
    else:
        if len(pesel) != 11:
            print(f"PESEL {pesel}: błędny format (długość {len(pesel)})")
        else:
            print(f"PESEL {pesel}: błędny format (zawiera niedozwolone znaki)")
```
