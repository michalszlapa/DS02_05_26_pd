```python
#Zadanie 2. Suma liczb dodatnich
liczby = [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3]
suma = 0
ile = 0
for temp in liczby:
    if temp > 0:
        suma = suma + temp
        ile = ile + 1
print("Suma dodatnich liczb:", suma)
print("Liczba dodatnich liczb:", ile)
```

```python
#Zadanie 5. Suma przez while
suma = 0
liczba = 1
while liczba <= 50:
    suma = suma + liczba
    liczba = liczba + 1
print("Suma liczb od 1 do 50:", suma)
# Sprawdzenie wzorem
n = 50
wynik_wzorem = n * (n + 1) / 2
print("Wynik ze wzoru:", wynik_wzorem)
```

```python
#Zadanie 12. Walidacja numeru PESEL (uproszczona)
pesels = ['92041812345', '8501015432', 'abc123', '78121698765']
for pesel in pesels:
    if len(pesel) == 11 and pesel.isdigit():
        print("PESEL", pesel + ": poprawny format")
    else:
        print("PESEL", pesel + ": błędny format")
```
