```python
liczby = [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3]

suma = 0
ile = 0

for temp in liczby:
    if temp > 0:
        suma += temp
        ile += 1

print("Suma liczb dodatnich:", suma)
print("Liczba liczb dodatnich:", ile)
```

```python
suma = 0
i = 1

while i <= 50:
    suma += i
    i += 1

print("Suma liczb od 1 do 50:", suma)

n = 50
print("Wynik ze wzoru:", n * (n + 1) / 2)
```

```python
pesele = ['92041812345', '8501015432', 'abc123', '78121698765']

for pesel in pesele:
    if len(pesel) == 11 and pesel.isdigit():
        print(f"PESEL {pesel}: poprawny format")
    else:
        print(f"PESEL {pesel}: błędny format")
```
