Zadanie 2. Suma liczb dodatnich
Dla listy [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3] oblicz sumę dodatnich wartości (pętla for , if temp >
0 ) oraz ile ich jest. Wypisz oba wyniki.

```python
lista = [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3]

dodatnie = [temp for temp in lista if temp > 0]

suma_dodatnich = sum(dodatnie)
licznik_dodatnich = len(dodatnie)

print("Suma dodatnich wartości:", suma_dodatnich)
print("Liczba dodatnich wartości:", licznik_dodatnich)
```

Zadanie 5. Suma przez while
Używając pętli while (nie for !) oblicz sumę liczb od 1 do 50. Sprawdź wynik wzorem n*(n+1)/2 =
1275 .

```python
n = 50
suma = 0
i = 1

while i <= n:
    suma += i
    i += 1

print(f"Suma liczb od 1 do {n}: {suma}") 
print(f"Wzór n*(n+1)/2 = {n * (n + 1) // 2}")
```

Zadanie 12. Walidacja numeru PESEL (uproszczona)
Dla listy PESEL-i ['92041812345', '8501015432', 'abc123', '78121698765'] :
Dla każdego numeru sprawdź:
długość dokładnie 11 znaków,
wszystkie znaki cyfrowe ( s.isdigit() – metoda z L5.
Wypisz raport: PESEL 92041812345: poprawny format albo PESEL 8501015432: błędny format (długość
10) .
Oczekiwany wynik: tylko pierwszy i ostatni są poprawne.

```python
lista = ['92041812345', '8501015432', 'abc123', '78121698765'] 

for pesel in lista:
    if len(pesel) == 11 and pesel.isdigit():
        print(f"PESEL '{pesel}' ma poprawny format.")
    else:
        print(f"PESEL '{pesel}' ma błędny format.")
```
